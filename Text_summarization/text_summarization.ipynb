```python
from langchain.chains import LLMChain
from langchain import PromptTemplate
generictemplate="""
write a summary of the speech:
Speech:{speech}
Translate the precise summary into {language}
"""

prompt=PromptTemplate(
    input_variables=['speech','language'],
    template=generictemplate
)
prompt
```

```python
complete_prompt=prompt.format(speech=speech,language="Malayalam")
```

```python
llm_chain=LLMChain(llm=llm,prompt=prompt)
summary=llm_chain.run({'speech':speech,'language':'Malayalam'})
summary
```

```python
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader('apjspeech.pdf')
docs=loader.load_and_split()
docs
```

```python
templates=""" 
Write a concise and short summary of the following speech.
speech:{text}
"""
prompt=PromptTemplate(
    input_variables=['text'],
    template=templates
)
```

```python
from langchain.chains.summarize import load_summarize_chain
chain=load_summarize_chain(llm,chain_type='stuff',prompt=prompt,verbose=True)
```

```python
Output_summary=chain.run(docs)
Output_summary
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter


text_splitter= RecursiveCharacterTextSplitter(chunk_size=2000,chunk_overlap=100).split_documents(docs)
text_splitter
```

```python
chunk_prompt=""" 
please summarize the below speech:
Speech:'{text}'
Summary:
"""

map_prompt_template=PromptTemplate(
    input_variables=['text'],
    template=chunk_prompt
)
```

```python
final_prompts=""" 
provide a  final summary of the entire speech with those important points.
Add a motivation Title,Start the precise summary with an introduction and provide the summary in number 
pionts for the speech.
Speech:{text} 
"""

final_prompt_templates=PromptTemplate(
    input_variable=['text'],
    template=final_prompts
)
```

```python
summary_chain=load_summarize_chain(
    llm=llm,
    chain_type="map_reduce",
    map_prompt=map_prompt_template,
    combine_prompt=final_prompt_templates,
    verbose=True
)
output=summary_chain.run(text_splitter)
```

```python
chain=load_summarize_chain(
    llm=llm,
    chain_type="refine",
    verbose=True
)
Output_summary=chain.run(text_splitter)
Output_summary
```
